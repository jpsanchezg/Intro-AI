### 3.2 Descripción de Datos

En nuestro primer set de datos [5], este contiene el tipo de personalidad del usuario mas su input, todo obtenido de los comentarios de un cuestionario hecho en una página de red social (Reddit). Aún así, los datos originales se pueden encontrar en [7].

#### 3.2.1 Métricas
Los tipos de datos que tenemos en nuestra posesión no necesitan un análisis de clustering, debido a que ya estan clasificados directamente a través de las identificaciones de personalidad de cada usuario. Esto significa que usaríamos clasificación para la construcción de nuestor modelo de predicción. Esto resulta en las siguientes métricas.

**Precisión**

La precisión es una métrica importante en nuestro análisis a través de una matriz de confusión. Debido a que nuestra matriz tiene varias celdas, a través de la precisión podemos analizar el porcentaje de predicciones positivas correctas por cada personalidad. Esto nos puede decir cual tipo de personalidad mejor predice nuestro sistema.
**Exactitud**

Ademas de la precisión, se usará la métrica de exactitud para definir el porcentaje de predicciones correctas hechas en el sistema en general. Esto resulta en una métrica de exito, y a través de multiples pruebas, puede decidir que tan estable es nuestro modelo.
